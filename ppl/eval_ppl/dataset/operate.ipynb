```python
import json

with open("time_related_queries.json") as f:
    content = json.load(f)
    
print(content[0])

# for sample in content:
#     sample["now"] = "TIMESTAMP('2025-05-27 08:53:46')"
for sample in content:
    if sample["target_index"] == "otel_logs":
        sample["now"] = "TIMESTAMP('2025-10-23 07:43:30')"
print(content[0])

with open("time_related_queries.json","w") as f:
    json.dump(content, f, indent=4)
```

```python
date_fields = {
    'sso_log': ['@timestamp', 'observerTime'],
    'flight': ['timestamp'],
    'otel_logs': ['endTime', 'events.time', 'startTime', 'traceGroupFields.endTime'],
    'log_index': ['timestamp', 'utc_time'],
    'ecommerce': ['customer_birth_date', 'order_date', 'products.created_on']
}

with open("time_related_queries.json") as f:
    content = json.load(f)
    
print(content[0])

for sample in content:
    fields = [field for field in date_fields[sample['target_index']] if field in sample['query']]
    assert len(fields) == 1
    field = fields[0]
    sample["date_field"] = field
    
with open("time_related_queries.json","w") as f:
    json.dump(content, f, indent=4)
```
